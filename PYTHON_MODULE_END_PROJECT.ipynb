# 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
team_count=data_players.groupby('Team').size().reset_index(name='No of Employees')
total_employees=len(data_players)
team_count['percentage split']=(team_count['No of Employees']/total_employees)*100
print(team_count)
```

# 2.Segregate the employees w.r.t different positions.

```python
position=data_players.groupby('Position')
for i,group_data in position:
    print("Employees in Position:",i)
    print(group_data)
```

# 3.Find from which age group most of the employees belong to.

```python
age_group=pd.cut(data_players['Age'],bins=[15,20,25,30,35,40],labels=['15-19','20-24','25-29','30-34','35-39'])
age_group_count=age_group.value_counts().sort_index()
plt.bar(age_group_count.index, age_group_count.values)
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.title('Distribution of Employees based on Age Group')
plt.show()
```

```python
age_range=pd.DataFrame(data_players['Age'])
bins = [15, 20, 25, 30, 35,40]
labels = ['15-19','20-24','25-29','30-34','35-39']
age_range['agerange'] = pd.cut(age_range.Age, bins, labels = labels,include_lowest = False)
print("Most of the employees belong to the age group of",age_range['agerange'].value_counts().index[0])
```

# 4.Find out under which team and position, spending in terms of salary is high.

```python
max=data_players.Salary.max()
spend=data_players[data_players['Salary']==max]
print("The spending in terms of salary is high for the team '{}' and the position '{}'.".format(spend.Team.iloc[0],spend.Position.iloc[0]))
```

# 5.Find if there is any correlation between age and salary , represent it visually.

```python
corr_data1=data_players.drop(['Name', 'Team', 'Position','College', 'Height','Weight','Number'], axis='columns')
corr = corr_data1.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix ")
plt.show()
```
