# 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
team_count=df.groupby('Team').size().reset_index(name='No of Employees')
total_employees=len(df)
team_count['percentage split']=(team_count['No of Employees']/total_employees)*100
print(team_count)
```

# 2.Segregate the employees w.r.t different positions.

```python
position=df.groupby('Position')
for i,group_data in position:
    print("Employees in Position:",i)
    print(group_data)
```

# 3.Find from which age group most of the employees belong to.

```python
age_group=pd.cut(df['Age'],bins=[20,30,40,50,60],labels=['20-29','30-39','40-49','50-59'])
age_group_count=age_group.value_counts().sort_index()
plt.bar(age_group_count.index, age_group_count.values)
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.title('Distribution of Employees based on Age Group')
plt.show()
```
